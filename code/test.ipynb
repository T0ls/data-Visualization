```python
import pandas as pd
```

```python
df = pd.read_csv('../data/csv/test.csv')
df.head()
```

```python
arrivals_per_year = df[(df['COUNTRY_RES_GUESTS'] == 'WORLD') & (df['TYPE_ACCOMMODATION'] == 'ALL') & (df['DATA_TYPE'] == 'AR') & (df['TIME_PERIOD'].apply(len) <= 4) & (df['LOCALITY_TYPE'] == 'ALL')]
arrivals_per_year[['OBS_VALUE', 'TIME_PERIOD']]
```

```python
arrivals = arrivals_per_year[['OBS_VALUE', 'TIME_PERIOD']].set_index('TIME_PERIOD')
arrivals.plot()
arrivals.plot.bar()
```
