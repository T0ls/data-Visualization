```python
from PIL import Image
from wordcloud import WordCloud
import numpy as np
import matplotlib.pyplot as plt
```

```python
txt = open("./wordcloud.txt", encoding='utf-8').read()
img = np.array(Image.open("./wordcloud.png"))

plt.figure(figsize = (16,9))

wordcloud = WordCloud(mask=img, background_color="GhostWhite", colormap="tab10", font_path="arial", font_step=1, min_word_length=4)
plt.imshow(wordcloud.generate(txt))

plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
